```python
import pandas as pd
```

Explore The Dataset

```python
df = pd.read_csv('spam.csv')
df
```

```python
df.groupby('Category').describe()
```

```python
df['Message'].nunique()
```

Labeling The Category

```python
from sklearn.preprocessing import LabelEncoder
```

```python
Label = LabelEncoder()
df['Category'] = Label.fit_transform(df['Category'])
```

```python
df
```

Count Vectorizor Technique

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
v = CountVectorizer()
X = v.fit_transform(df['Message'])
```

```python
X.toarray()
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,df.Category,test_size=0.2)
```

Model Selection

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
model.fit(X_train,y_train)
```

```python
email = [
    'Hey Rafay, Can we get together',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]
email_count = v.transform(email)
model.predict(email_count)
```

```python
model.score(X_test,y_test)
```

Now using Pipeline to easy our work

```python
from sklearn.pipeline import Pipeline
```

```python
clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB()),
])
```

```python
X = df['Message']
y = df.Category
```

```python
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.2)
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
clf.predict(email)
```
