Multivariate Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('hiring.csv')
df
```

```python
df['experience'] = df['experience'].fillna('five')
```

```python
df
```

```python
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(df['test_score(out of 10)'].mean())
```

```python
df
```

#### library used to convert number text to integer value is word2number

```python
from word2number import w2n
```

```python
df['experience'] = df['experience'].apply(lambda x: w2n.word_to_num(x))
```

```python
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['experience','test_score(out of 10)','interview_score(out of 10)']], df['salary($)'])
```

```python
plt.scatter(df['experience'], df['salary($)'], label = 'exp')
plt.scatter(df['test_score(out of 10)'], df['salary($)'], label = 'test')
plt.scatter(df['interview_score(out of 10)'], df['salary($)'], label = 'interview')

plt.legend()

plt.show()
```

```python
reg.predict([[2,9,6]])
```

```python
reg.predict([[12,10,10]])
```

```python
reg.score(df[['experience','test_score(out of 10)','interview_score(out of 10)']], df['salary($)'])
```
